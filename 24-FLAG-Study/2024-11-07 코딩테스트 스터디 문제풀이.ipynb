## 1. 2292번 풀이
이 문제에서 숫자가 어느 층에 들어갈지 알아내는 문제입니다. 각 층에 들어갈 숫자들를 알면 문제가 해결될거 같다고 생각했습니다.그래서 1층부터 계산을 해보니 6의 배수만큼 더해지는 규칙을 발견할 수 있었습니다. 이제 이 규칙을 이용하여 1층부터 탐색을 진행합니다. 1층의 마지막 숫자는 1입니다. 이를 변수 last_num 에 넣고 다음 층의 마지막 숫자에 더해질 6의 배수는 6*1 이므로 count 는 1부터 시작합니다. 반복문의 조건을 알고자하는 숫자 N 보다 층의 마지막 숫자 last_num 이 크거나 같아질 때까지 반복합니다. 그때의 count 가 현재 층을 가리키고 있으므로 결과로 count 를 출력합니다.

1층 $1+6*0=6$

2층 ~$1+6*1=7$

3층 ~$1+6*1+6*2=19$

...

N층 ~$1+\sum_{i=0}^{n-1} 6i$

```python
N = int(input())
count = 1
last_num = 1
while last_num < N:
    last_num += 6*count
    count += 1
print(count)
```

## 2. 2581번 풀이

### 단순구현 방법

```python
M = int(input())
N = int(input())
nums = list(range(M,N+1))
res = [] # 소수들이 담길 리스트
for num in nums:
    if num == 1:
        continue

    # num 소수인지 판별
    is_prime = True
    for i in range(2,num):
        if num % i == 0: # num은 i의 배수이므로 소수가 아님!
            is_prime = False
            break

    if is_prime: # 소수이면 리스트에 추가
        res.append(num)

if len(res) != 0:
    print(sum(res),min(res),sep='\n')
else:
    print(-1)
```

### 에라토스테네스의 체 이용하는 방법

```python
M = int(input())
N = int(input())

res=set(range(2, N+1)) # 2부터 N을 갖는 집합 자료형 선언
for i in range(2, N+1): # 2부터 N까지 반복
    if i in res:
        res -= set(range(2*i, N+1, i)) # 차집합 연산을 통해 자신을 제외한 소수의 배수 제거

# res는 2~N 내의 모든 소수가 담김 -> M보다 크거나 같은 수만 얻어서 결과를 계산한다
res=list(filter(lambda x:x>=M,res)) # filter(조건 함수, iterable한 객체) -> filter 객체 =(리스트로 캐스팅)> list
if len(res) != 0: # 소수가 있음
    print(sum(res),min(res),sep='\n')
else: # 소수가 없음
    print(-1)
```

## 3. 5073번 풀이

```python
import sys
while True:
    angles = list(map(int,sys.stdin.readline().strip().split()))
    if sum(angles) == 0: # 0 0 0 의 합은 0이므로 종료!
        break
    num_of_unique = len(set(angles)) # set(집합) 자료구조를 이용해서 중복되지 않는 숫자의 개수를 구함
    if sum(angles) - 2*max(angles) <= 0: # 삼각형 성립 조건 체크 : 가장 긴 변의 길이보다 나머지 두변의 길이가 더 길어야 함
        print('Invalid')
    elif num_of_unique == 3: # 삼각형이 성립하면서 모두 다른 길이라면 일반 삼각형
        print('Scalene')
    elif num_of_unique == 2: # 삼각형이 성립하면서 두 변이 같다면 이등변 삼각형
        print('Isosceles')
    else: # 삼각형이 성립하면서 세 변이 같다면 정삼각형
        print('Equilateral')
```
